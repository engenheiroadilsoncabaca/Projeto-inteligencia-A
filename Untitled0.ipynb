 Pré-processamento

```python
import pandas as pd
import numpy as np

LABEL_COLS = ['falha_1', 'falha_2', 'falha_3', 'falha_4', 'falha_5', 'falha_6', 'falha_outros']

def load_data(filepath):
    df = pd.read_csv(filepath)
    df = clean_labels(df)
    return df

def clean_labels(df):
    for col in LABEL_COLS:
        df[col] = df[col].astype(str).str.lower().str.strip()
        df[col] = df[col].replace({'true': 1, 'false': 0, 'sim': 1, 'não': 0, 'nao': 0})
        df[col] = df[col].where(df[col].isin([0, 1]), 0)  # substitui valores inválidos por 0
        df[col] = df[col].fillna(0).astype(int)
    return df

def preprocess_features(df):
    return df.drop(columns=['id'] + LABEL_COLS)

def get_labels(df):
    return df[LABEL_COLS]

# Carrega o dataset
df = load_data("bootcamp_train.csv")
```

EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_class_distribution(df):
    counts = df[LABEL_COLS].sum().sort_values(ascending=False)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=counts.index, y=counts.values)
    plt.title("Distribuição de falhas")
    plt.xlabel("Tipo de falha")
    plt.ylabel("Número de ocorrências")
    plt.tight_layout()
    plt.show()

def plot_correlation_matrix(df):
    df_numeric = df.select_dtypes(include=['number'])  # apenas colunas numéricas
    corr = df_numeric.corr()
    plt.figure(figsize=(12, 8))
    sns.heatmap(corr, cmap="coolwarm", annot=False)
    plt.title("Matriz de Correlação entre Variáveis Numéricas")
    plt.tight_layout()
    plt.show()

# Execução com foco em features
features = preprocess_features(df)
plot_class_distribution(df)         # distribuição de rótulos
plot_correlation_matrix(features)   # correlação entre features
```
